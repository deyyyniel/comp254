```python
import time
import matplotlib.pyplot as plt
import numpy as np


def prefix_average1(S):
    n = len(S)
    A = [0] * n
    for j in range(n):
        total = 0
        for i in range(j + 1):
            total += S[i]
        A[j] = total / (j + 1)
    return A


def prefix_average2(S):
    n = len(S)
    A = [0] * n
    for j in range(n):
        A[j] = sum(S[0 : j + 1]) / (j + 1)
    return A


# Generate input sizes
input_sizes = [100, 200, 400, 800, 1600, 3200, 6400]

# Store execution times
times1 = []
times2 = []

for size in input_sizes:
    S = np.random.rand(size).tolist()

    # Time for prefix_average1
    start_time = time.time()
    prefix_average1(S)
    end_time = time.time()
    times1.append(end_time - start_time)

    # Time for prefix_average2
    start_time = time.time()
    prefix_average2(S)
    end_time = time.time()
    times2.append(end_time - start_time)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, times1, label="prefix_average1", marker="o")
plt.plot(input_sizes, times2, label="prefix_average2", marker="o")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time vs Input Size for Prefix Average Algorithms")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import time
import random
import statistics

def unique1(S):
    for j in range(len(S)):
        for k in range(j + 1, len(S)):
            if S[j] == S[k]:
                return False
    return True

def measure_time(n, iterations=10):
    times = []
    for _ in range(iterations):
        S = random.sample(range(2 * n), n)  # Generate a list of n unique integers
        start_time = time.time()
        unique1(S)
        end_time = time.time()
        times.append(end_time - start_time)
    return statistics.mean(times), statistics.stdev(times)

elapsed_time, std_dev = measure_time(5030)
print(f"Average elapsed time: {elapsed_time:.6f} seconds")
print(f"Standard deviation: {std_dev:.6f} seconds")
```

```python
def find_missing_number(arr):
    n = len(arr) + 1
    total_sum = n * (n - 1) // 2
    array_sum = sum(arr)
    return total_sum - array_sum

def main():
    # Test case 1: missing number is 2
    arr1 = [0, 1, 3, 4]
    print(f"Missing number in {arr1} is {find_missing_number(arr1)}")

    # Test case 2: missing number is 4
    arr2 = [0, 1, 2, 3, 5, 6, 7, 8, 9]
    print(f"Missing number in {arr2} is {find_missing_number(arr2)}")

    # Test case 3: missing number is 0
    arr3 = [1, 2, 3, 4, 5]
    print(f"Missing number in {arr3} is {find_missing_number(arr3)}")

    # Test case 4: missing number is 9
    arr4 = [0, 1, 2, 3, 4, 5, 6, 7, 8]
    print(f"Missing number in {arr4} is {find_missing_number(arr4)}")

if __name__ == "__main__":
    main()
```
